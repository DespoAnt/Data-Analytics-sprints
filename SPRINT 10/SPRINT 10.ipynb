```python
def conteo_nulos_faltantes(df):
    """
    Cuenta los valores faltantes:
    - Nulos (NaN)
    - Espacios en blanco (' ') o guiones ('-')
    - Valores 99 (NS/NC)

    Da un diccionario con el nombre de la columna y el cada conteo por tipo
    """
    conteo = {}

    for col in df.columns:
        null_count = df[col].isnull().sum()
        empty_hyphen_count = ((df[col] == ' ') | (df[col] == '-')).sum()
        ns_nc_count = ((df[col] == 99)).sum()

        if null_count > 0 or empty_hyphen_count > 0 or ns_nc_count > 0:
            conteo[col] = {}
            if null_count > 0:
                conteo[col]['null'] = null_count
            if empty_hyphen_count > 0:
                conteo[col]['empty_hyphen'] = empty_hyphen_count
            if ns_nc_count > 0:
                conteo[col]['ns_nc'] = ns_nc_count

    return conteo
```

# Funciones

```python
def cambiar_binario(df, columnas=None):
    
    '''
    Cambiar columnas de un df a binarias según el mapeo
    en columna pasar una lista de columnas a cambiar o sino por defecto (None) se cambian todas
    mapeo: los valores de 1 a 6 cambian a 1 y los 7 a 0

    devuelve la copia del df original binarizada
    '''

    a_binario = {1: 1, 2: 1, 3: 1, 4: 1, 5: 1, 6: 1, 7: 0}

    df_binario = df.copy()

    if columnas is None:
        columnas = df.columns

    df_binario[columnas] = df[columnas].replace(a_binario)

    return df_binario
            
```

```python
def binarizar_columna(df, columna, sufijo='_original', mapeo=None):
       
    '''
    Copiar una columna y cambiar sus valores por binarios según el mapeo
    en columna el nombre de la columna a cambiar
    en sufijo el sufijo para guardar la copia original, por defecto '_original'
    mapeo: los valores de 1 a 6 cambian a 1 y los 7 a 0

    devuelve el df con la columna binarizada y la copia de la original
    '''
    
    if mapeo is None:
        mapeo = {1: 1, 2: 1, 3: 1, 4: 1, 5: 1, 6: 1, 7: 0}

    # Copiar columna original
    copia_columna = f"{columna}{sufijo}"
    df[copia_columna] = df[columna].copy()

    # Reemplazar valores
    df[columna] = df[columna].replace(mapeo)

    return df
```

#### crear los grupos de categorias de actividades


```python
def crear_df_categorias(letras):

    #Una función para crear los grupos de las diferentes categorias eligiendo las columnas por la letra inicial
    
    cols_comunes = ['ID', 'EDAT', 'AmbitTerritorial', 'SEXE']

    if isinstance(letras, str):
        letras = [letras]

    cols_adicionales = [col for col in data2024_1.columns if any(col.lower().startswith(letra.lower()) for letra in letras)]

    cols_final =cols_comunes + cols_adicionales
    
    return data2024_1[cols_final]
```

#### función para formatos de actividades

```python
def formatos_actividades(grupo, columna, col_formatos):

    ''' 
    crear un df con las columnas de los formatos por actividad de cada grupo de activitad
    en grupo el grupo de actividad (creado para cada una con las columnas filtradas)
    en columna la columna index del crosstab (SEXE, EDAT, AmbitTerriorial)
    como col_formatos, el grup de columnes de cada formato'''

    resultados = {}

    for col in col_formatos:
        tabla_ccross = round(pd.crosstab(grupo[columna], grupo[col] == 1, normalize='index') * 100, 2)
    # Tomamos solo la columna True (donde == 1)
        resultados[col] = tabla_ccross.get(True, pd.Series(0, index=tabla_ccross.index))

    return pd.DataFrame(resultados)
```

```python
# crear un df con las columnas de las activiades de barreras sobre no participación

etiquetas_motivos = {
    '1': 'Falta de Tiempo', '2': 'Responsabilidades Familiares', '3': 'Falta de Interés, prefiero otras cosas',
    '4': 'Falta de información, no tengo conocimientos', '5': 'Problemas de salud', '6': 'Falta de acompañantes',
    '7': 'Falta de oferta (o oferta de baja calidad)', '8': 'Precio', '9': 'Dificultades de accesibilidad física',
    '10': 'Dificultades de accesibilidad digital', '95': 'Otros motivos'
}

iniciales_columnas = ['M7#', 'VJ2#', 'ES6#', 'LL11#', 'EX7#', 'CN6#', 'CO7#']

#codigos de impedimentos del 1 a 10, y 95

motivos_codigos = [str(i) for i in range(1, 11)] + ['95'] 

columnas_motivos = []
for inicial in iniciales_columnas:
    for codigo in motivos_codigos:
        columnas_motivos.append(inicial + codigo)

columnas_motivos_existentes = [col for col in columnas_motivos if col in data2024_1.columns]

# hacer el melt del df para poder realizar los cálculos
df_largo = pd.melt(data2024_1, value_vars=columnas_motivos_existentes,var_name='columna_motivo',value_name='respuesta') 

# separar el codigo del motivo que aparece después de #
df_largo['codigo_motivo'] = df_largo['columna_motivo'].str.split('#').str[-1]

df_solo_si = df_largo.query('respuesta == 1')

df_conc = pd.DataFrame(index=data2024_1.index)
for codigo, etiqueta in etiquetas_motivos.items():
    cols_motivo = [col for col in columnas_motivos_existentes if col.endswith(f'#{codigo}')]
    if cols_motivo:
        df_conc[etiqueta] = data2024_1[cols_motivo].max(axis=1)

data_final = []
for columna in df_conc.columns:
    conteo_personas = df_conc[columna].sum()
    
    porcentaje_barrera = (conteo_personas / total_muestra) * 100
    
    data_final.append({
        'Motivo': columna,
        'Conteo_Personas': conteo_personas,
        'Porcentaje_barrera': round(porcentaje_barrera, 1)
    })

df_ranking_final = pd.DataFrame(data_final)
df_ranking_final = df_ranking_final.sort_values(by='Porcentaje_barrera', ascending=False)
```
